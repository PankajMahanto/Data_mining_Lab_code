```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# For visualization
from sklearn.decomposition import PCA
```

```python
# Load the dataset
data = pd.read_csv('Mall_Customers.csv')

# Select relevant features
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
# Suppose from dendrogram we decide 5 clusters
labels_hierarchical = fcluster(linked, 5, criterion='maxclust')

# Add to dataframe
data['Cluster_Hierarchical'] = labels_hierarchical
```

```python
# Reduce to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot Hierarchical Clustering
plt.figure(figsize=(8,5))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=data['Cluster_Hierarchical'], palette='Set1')
plt.title('Hierarchical Clustering Results (PCA reduced)')
plt.show()

# Plot DBSCAN Clustering
plt.figure(figsize=(8,5))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=data['Cluster_DBSCAN'], palette='Set2')
plt.title('DBSCAN Clustering Results (PCA reduced)')
plt.show()
```
