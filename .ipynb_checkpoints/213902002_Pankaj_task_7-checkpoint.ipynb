## Working on WINE dataset

```python
from sklearn.datasets import load_wine
```

### Loading data

```python
wine_dt = load_wine()
```

```python
wine_dt
```

```python
wine = wine_dt.data
wine
```

### Dataframe convert

```python
wine = pd.DataFrame(wine_dt.data,columns=wine_dt.feature_names)
```

```python
wine.head()
```

### Feature names

```python
wine_dt.feature_names
```

### Target names

```python
wine_dt.target_names
```

### Data divided input and output

```python
x = wine_dt.data
y = wine_dt.target
x.shape,y.shape
```

### Splitting data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.20,random_state=42)
```

```python
xtrain.shape,ytrain.shape
```

## Model Trainning Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion = 'entropy')
```

```python
dt.fit(xtrain,ytrain)
```

```python
y_pred = dt.predict(xtest)
```

## Result analysis

```python
from sklearn.metrics import accuracy_score, mean_squared_error,confusion_matrix
```

### Accuracy

```python
print('Accuracy: ',accuracy_score(ytest,y_pred))
```

### Confution matrix

```python
print('Confusion Matrix: ',confusion_matrix(ytest,y_pred))
```

```python
confusion_matrix(ytest,y_pred)
```

## Decision Tree Generated

```python
from sklearn import tree
fig,ax = plt.subplots(figsize=(15,15))
tree.plot_tree(dt,ax=ax,feature_names=wine_dt.feature_names)
plt.show()
```

# KNN Model Define here

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(xtrain)
```

```python
xtrain = scaler.transform(xtrain)
```

```python
xtest = scaler.transform(xtest)
```

## Model initialized

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(xtrain,ytrain)
```

```python
y_pred= knn.predict(xtest)
```

## Result Analysis

### Accuracy

```python
accuracy_score(ytest,y_pred)
```

### Confusion Matrix

```python
confusion_matrix(ytest,y_pred)
```

## Calculating error to get optimized k values

```python
error=[]
max_k = len(xtrain)
for i in range(1,max_k+1):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_i = knn.predict(xtest)
    error.append(np.mean(pred_i!=ytest))
    print(np.mean(pred_i!=ytest))


plt.figure(figsize=(12,6))

plt.plot(range(1,max_k+1),error,color='red',linestyle='dashed',marker='o',
        markerfacecolor='blue',markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K value')
plt.ylabel('Mean value')

plt.show()

```
