```python
x,y = make_blobs(n_samples=150, n_features=2,
                centers=3, cluster_std=0.5,
                shuffle=True, random_state=0)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(
    n_clusters=3,init = 'random',
    n_init=10, max_iter=300,
    tol=1e-04, random_state=0
)

y_km=km.fit_predict(x)
```

```python
y_km
```

```python
plt.scatter(
    x[y_km==0,0],x[y_km==0,1],
    s=50,c='lightgreen',
    marker='s',edgecolors='black',
    label='cluster 1'
)
plt.scatter(
    x[y_km==1,0],x[y_km==1,1],
    s=50,c='orange',
    marker='s',edgecolors='black',
    label='cluster 2'
)
plt.scatter(
    x[y_km==2,0],x[y_km==2,1],
    s=50,c='lightblue',
    marker='s',edgecolors='black',
    label='cluster 3'
)

plt.scatter(
    km.cluster_centers_[:,0],km.cluster_centers_[:,1],
    s=250,marker='*',
    c='red',edgecolors='black',
    label='centroids'
)


plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

# Step 1: Generate Synthetic Multivariate Data

```python
from sklearn.preprocessing import StandardScaler
def generate_data(samples=300, features=2, centers=4, random_state=42):
    data,labels = make_blobs(n_samples=samples,n_features=features,centers = centers,
                            random_state=random_state)
    data = StandardScaler().fit_transform(data)

    return data,labels
```

# Step 2: Apply K-Means Clustering

```python
def apply_kmeans(data, k=4):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    return kmeans.labels_, kmeans.cluster_centers_
```

# Step 3: Visualize Clusters

```python
def plot_clusters(data, labels, centers, title="K-Means Clustering"):
    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1], c=labels, s=50, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True)
    plt.show()
```

# ==== Usage Example ====
# Change number of clusters or samples here

```python

data, true_labels = generate_data(samples=500, features=2, centers=5)
pred_labels, centers = apply_kmeans(data, k=5)
plot_clusters(data, pred_labels, centers, title="K-Means with 5 Clusters")
```
