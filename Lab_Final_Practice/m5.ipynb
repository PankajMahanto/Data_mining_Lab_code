```python
from mlxtend import frequent_patterns as fp
dir(fp)
```

```python
from mlxtend.frequent_patterns import apriori as ap
from mlxtend.frequent_patterns import association_rules as ar
```

```python
u_c
```

```python
bask_f = (dt[dt['Country']=='France']
         .groupby(['InvoiceNo','Description'])['Quantity']
         .sum().unstack().reset_index().fillna(0)ltw1
         .set_index('InvoiceNo')
         )
```

## One hot Encoding

```python
def One(P):
    if P<=0:
        return 0
    if P>=1:
        return 1
```

```python
bask_f_e = bask_f.applymap(One)
bask_f = bask_f_e
bask_f
```

## Build the model and Analysis Results

```python
frq_item1 = ap(bask_f,min_support=0.05,use_colnames=True)
frq_item1
```

```python
rules1 = ar(frq_item1,metric='lift',min_threshold=1)
rules1 = rules1.sort_values(['confidence','lift'],ascending=[False,False])
print(rules1.head())
```
